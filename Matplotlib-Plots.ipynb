## FINDING LIMITS

```python
df["tip"].describe()
```

```python
p25=df["tip"].quantile(.25)
p25
```

```python
p75=df["tip"].quantile(.75)
p75
```

```python
IQR=p75-p25
IQR
```

```python
upper_limit=p75+1.5*IQR
lower_limit=p25-1.5*IQR
```

```python
upper_limit  ##  AFTER THIS LIMIT WE HAVE  UPPER OUTLIERS
```

```python
lower_limit   ###  CAN BE CONSIDERED AS LOWER OUTLIERS
            ### SO FROM GRAPH  ALL THA OUTLIER ARE ABOVE 5.90
```

```python
df[df["tip"]> upper_limit]
```

```python
len(df[df["tip"]> upper_limit])
```

```python
df[df["tip"]< lower_limit]
```

```python
len(df[df["tip"]<lower_limit])
```

## how to handle outliers
### 1: trimming => not recomended as info is lost
### 2: capping => data balance with upper limit or lower limit


## trimming

```python
dff=df[(df["tip"]<lower_limit) & (df["tip"]>upper_limit)]
```

```python
dff
```

```python
dff["tip"].skew()
```

```python
new_df_cap = df.copy()  ## create a copy for handling outlier in tips
new_df_cap

```

```python
new_df_cap['tip'] = np.where(new_df_cap['tip'] > upper_limit, upper_limit, 
                                              np.where(new_df_cap['tip'] < lower_limit, lower_limit,
                                                       new_df_cap['tip'])) # for middle part of box which can be replaced with any one 



plt.boxplot(new_df_cap['tip']);
```
